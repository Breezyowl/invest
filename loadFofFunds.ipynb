```python
#获得当前所有LOF基金,并将其存入列表
query_LOF=query(finance.FUND_MAIN_INFO).filter(finance.FUND_MAIN_INFO.operate_mode_id.in_(['401006','401005']) )
df=finance.run_query(query_LOF)[['main_code','name','advisor','operate_mode','underlying_asset_type_id','start_date']]
str_funds_list=df['main_code'].tolist()
#对基金进行选择操作，逻辑成交额大于10万，并且振幅大于3个点
quotation = eq.use('tencent') # 新浪 ['sina'] 腾讯 ['tencent', 'qq']
funds_dict = quotation.stocks(str_funds_list)

valid_funds = []
for key,value in funds_dict.items():
    if value['成交额(万)'] >= 100000 and value['振幅'] >3:
        valid_funds.append(key)


res = df[df['underlying_asset_type_id'].isin(valid_funds)]  
```
